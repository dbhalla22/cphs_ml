```python
import pandas as pd
df = pd.read_parquet('/content/drive/MyDrive/merged_file.parquet',
                     columns=['STATE', 'MONTH',
                              'M_EXP_FOOD', 'SIZE_GROUP', 'AGE_GROUP', 'GENDER_GROUP', 'EDU_GROUP',
                              'OCCUPATION_GROUP', 'REGION_TYPE', 'TOT_INC',
                              'INC_OF_HH_FRM_SELF_PRODN', 'INC_OF_HH_FRM_PVT_TRF'])
```

```python
import pandas as pd


# Split the 'MONTH' column into two new columns
df[['Month_Name', 'YEAR']] = df['MONTH'].str.split(' ', expand=True)

# Convert Year to integer if needed
df['YEAR'] = df['YEAR'].astype(int)

# Optional: Preview the result
print(df[['MONTH', 'Month_Name', 'YEAR']].head())
```

```python
# Filter for year 2023
df_2023 = df[df['YEAR'] == 2023]
```
